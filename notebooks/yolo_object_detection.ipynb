```python
# Import YOLO

from ultralytics import YOLO
```

```python
# Load OBB pretrained model

model = YOLO("yolov8s-obb.pt")
```

```python
# Mount Goolge Drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Set Dataset Path

DATASET_PATH = "/content/drive/MyDrive/yolo_dataset"
```

```python
# Train the Model

model.train(
    data=DATASET_PATH + "/data.yaml",  # path to your dataset YAML
    epochs=80,                         # can increase if validation loss not stable
    imgsz=512,                          # match your stretched preprocessing
    batch=32,                            # A100 GPU can handle large batch
    device=0,                            # 0 = default GPU
    patience=10,                         # early stopping patience
    workers=8,                           # data loading threads
    mosaic=1.0,                          # enable mosaic
    mixup=0.1,                           # small mixup for OBB
    fliplr=0.5,                          # horizontal flip probability
    hsv_h=0.11,                          # hue adjustment (¬±20¬∞ normalized)
    hsv_s=0.2,                           # saturation adjustment
    hsv_v=0.15,                          # brightness/exposure adjustment
    name="yolov8_obb_custom"             # name for the training run
)
```

```python
# Load best model from training
model = YOLO("runs/obb/yolov8_obb_custom/weights/best.pt")

# Run validation
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source=DATASET_PATH + "/val/images",  # path to validation images
    imgsz=512,
    conf=0.25,                            # confidence threshold
    save=True                             # saves predictions in runs/obb/predict
)
```
